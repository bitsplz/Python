# Trignometry functions and ploting them

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.sin(180)
np.sin(90)
```

```python
np.cos(180)
```

```python
np.tan(180)
```

```python
x_axis=np.arange(0,2*np.pi,0.1)
y_axis_sin=np.sin(x_axis)
plt.plot(x_axis,y_axis_sin)
plt.show()
```

```python
y_axis_cos=np.cos(x_axis)
plt.plot(x_axis,y_axis_cos)
plt.show()
```

```python
y_axis_tan=np.tan(x_axis)
plt.plot(x_axis,y_axis_tan)
plt.show()
```
